## Loc and iloc are used to identify subset of pandas dataframe
loc is label-based, which means that you have to specify rows and columns based on their row and column labels.  iloc is integer position-based, so you have to specify rows and columns by their integer position values (0-based integer position).

loc[row_label, column_label]
iloc[row_position, column_position]

```python
#Creat a dataset with two segment, half seg1 and half seg2, seg1 has 3% bad rate and seg2 has 4% bad rate
N=10000  #number of points
df = pd.DataFrame()
df['customer_id'] = np.array([i for i in range(N)])
df.loc[df.index<N/2,'segment'] = 'seg1'  #index and segment are labels

df.loc[df.index<N/2,'bad'] = [int(x<0.03) for x in np.random.uniform(0, 1, int(N/2))] 
df.loc[df.index>=N/2,'segment'] = 'seg2'
df.loc[df.index>=N/2,'bad'] = [int(x<0.04) for x in np.random.uniform(0, 1, int(N/2))] 
df.head()

```

```python
#convert a column into numpy array
df.bad.values
```

```python
#drop useless data for modeling
df2=df.drop(columns=['customer_id'])
df2
```

## Panda's own histogram chart

```python
#Use histogram to find distribution
fig = plt.figure(figsize = (15,20))
ax = fig.gca()
hist = df[['reduced_lunch','school_rating']].hist(ax=ax, by=df['school_rating'],bins=5)
```

```python
#Use scatter plot to find relationship 
df.plot.scatter(x='school_rating',y='reduced_lunch',c='Blue')
```

```python
fig, ax = plt.subplots(figsize=(14,8))

ax.set_ylabel('school_rating')

# boxplot with only these two variables
_ = df[['reduced_lunch', 'school_rating']].boxplot(by='school_rating', figsize=(13,8), vert=False, sym='b.', ax=ax)
```

```python
plt.figure(figsize=(14,8))
_ = sns.regplot(data=df, x='reduced_lunch', y='school_rating')
```

```python
# create tabular correlation matrix
corr = df.corr()
_, ax = plt.subplots(figsize=(13,10)) 

# graph correlation matrix
_ = sns.heatmap(corr, ax=ax,
                xticklabels=corr.columns.values,
                yticklabels=corr.columns.values,
                cmap='coolwarm')
```
